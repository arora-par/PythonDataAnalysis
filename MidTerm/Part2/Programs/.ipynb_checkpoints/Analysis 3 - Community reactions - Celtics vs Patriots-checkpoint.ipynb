### What did the fans had to say about these teams?

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import os
import glob
import calendar
import json
```

```python
# Folder locations
celtics_path = '../Data/NYT/Comments/Boston Celtics'
patriots_path = '../Data/NYT/Comments/New England Patriots'
```

```python
# Get all the words from the comments
celtics_comment_word_bag = ['Celtics']

for article_path in glob.glob(celtics_path + '/*.json'):
    with open(article_path, 'r', encoding="utf-8-sig") as f:     
        jsonobj2 = json.load(f)
        comment_body2 = jsonobj2['commentBody']
        celtics_comment_word_bag.extend(comment_body2.split())
            
patriots_comment_word_bag = ['Patriots']
for article_path in glob.glob(patriots_path + '/*.json'):
    with open(article_path, 'r', encoding="utf-8-sig") as f:     
        jsonobj1 = json.load(f)
        comment_body1 = jsonobj1['commentBody']
        patriots_comment_word_bag.extend(comment_body1.split())
                        
```

```python
# Plot word clouds for both teams from the published articles
celtics_stopwords = [ 'Boston', 'points', 'game', 'night', 'th', 'tr', 'New', 'br']
celtics_stopwords.extend(STOPWORDS)
wordcloud = WordCloud(stopwords=celtics_stopwords,
                          background_color='white',
                          max_words=200,
                          width=1200,
                          height=1000
                         ).generate(' '.join(celtics_comment_word_bag))


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python

patriots_stopwords = [ 'Boston', 'points', 'game', 'night', 'th', 'tr', 'New', 'br', 'will']
patriots_stopwords.extend(STOPWORDS)
wordcloud = WordCloud(stopwords=patriots_stopwords,
                          background_color='white',
                          max_words=200,
                          width=1200,
                          height=1000
                         ).generate(' '.join(patriots_comment_word_bag))


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
