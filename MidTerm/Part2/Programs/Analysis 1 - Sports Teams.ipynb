## Year 2016 - Boston Celtics vs New England Patriots popularity

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import os
import glob
import calendar
import json
```

### Which team was published more often?

```python
# Folder locations
celtics_path = '../Data/NYT/Sports/Boston Celtics'
patriots_path = '../Data/NYT/Sports/New England Patriots'

# Key,Value pairs of month,number to help with sorting
month_name = {v: k for k,v in enumerate(calendar.month_abbr)}

monthly_pubs_celtics = []
monthly_pubs_patriots = []

# Read no of files from the corresponding Data folders
for monthly_path in glob.glob(celtics_path + '/*'):
    monthly_pubs_celtics.append((month_name[monthly_path[-3:]], len(os.listdir(monthly_path)) - 1))
    
for monthly_path in glob.glob(patriots_path + '/*'):
    monthly_pubs_patriots.append((month_name[monthly_path[-3:]], len(os.listdir(monthly_path)) - 1))

monthly_pubs_celtics.append((0,0))
monthly_pubs_patriots.append((0,0))

monthly_pubs_celtics = sorted(monthly_pubs_celtics)    
monthly_pubs_patriots = sorted(monthly_pubs_patriots)    

# Plot a monthly track record for both the teams
plt.plot([v for k,v in monthly_pubs_celtics], label ='Boston Celtics')
plt.plot([v for k,v in monthly_pubs_patriots], label ='New England Patriots')
plt.axes().set_ylim(0,170)
plt.axes().patch.set_facecolor('#FFFFFF')
plt.axes().grid(b=True, which='major', color='#DDDDDD', linestyle='-')
plt.axes().set_title("Articles Published (NYT API - Article Search API)",fontweight='bold',fontsize=15,position=[.5,1.05])
plt.axes().set_ylabel("No. of articles",fontsize=12,fontstyle='italic')
plt.axes().set_xlabel("Month",fontsize=12,fontstyle='italic')
plt.xticks([i for i in range(13)])
plt.legend(loc='upper right')
plt.show()


```

### How often did they win/lose?

```python
# Get all the important words from the articles
word_bag = []
for monthly_path in glob.glob(celtics_path + '/*'):
    for article_path in glob.glob(monthly_path + '/*.json'):
        with open(article_path, 'r', encoding="utf-8-sig") as f:     
            jsonobj = json.load(f)
            para = jsonobj['lead_paragraph']
            word_bag.extend(para.split())

word_bag_patriots = []
for monthly_path in glob.glob(patriots_path + '/*'):
    for article_path in glob.glob(monthly_path + '/*.json'):
        with open(article_path, 'r', encoding="utf-8-sig") as f:     
            jsonobj = json.load(f)
            para = jsonobj['lead_paragraph']
            if (para):
                word_bag_patriots.extend(para.split())
            else:
                main_head = jsonobj['headline']['main'] 
                word_bag_patriots.extend(main_head.split())
                
# Lemmatization - count occurrences of words 'win' 'lose'              
            
```

### What did the media had to say about these teams?

```python
# Plot word clouds for both teams from the published articles
celtic_stopwords = [ 'Boston', 'points', 'game', 'night', 'th', 'tr', 'New']
celtic_stopwords.extend(STOPWORDS)
wordcloud = WordCloud(stopwords=celtic_stopwords,
                          background_color='white',
                          width=1200,
                          height=1000
                         ).generate(' '.join(word_bag))


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# Plot word clouds for both teams from the published articles

patriots_stopwords = [ 'England', 'points', 'game', 'night', 'th', 'tr', 'New']
patriots_stopwords.extend(STOPWORDS)
wordcloud = WordCloud(stopwords=patriots_stopwords,
                          background_color='white',
                          width=1200,
                          height=1000
                         ).generate(' '.join(word_bag_patriots))


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
